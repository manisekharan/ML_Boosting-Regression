```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import lightgbm as lgb
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
regressor = lgb.LGBMRegressor()
regressor.fit(X_train, y_train)
```

```python
#regressor.score(X_train, y_train)

y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#Import pickle lib to save the model
import pickle
filename = "finalized_model_ada_boosting_insurance_charges.sav"
#Save the model to file using write/binary mode
pickle.dump(regressor, open(filename,'wb'))
#Load the model from the file and test a prediction for a input
loaded_model = pickle.load(open("finalized_model_ada_boosting_insurance_charges.sav",'rb'))
result = loaded_model.predict([[5,30.200,1,1,0]])
result
```
